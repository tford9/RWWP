## Erdos Renyi Graph Generation

```python
scaling_data = pd.DataFrame(columns=['TYPE', 'ORDER', 'SIZE', 'WALL', 'CPU', 'RESULT'])
for x in range(0,2):
    for i in range(3,10,1):
        G = nx.erdos_renyi_graph(2**i, 10**(-1), seed=None, directed=True)
        nx.write_edgelist(G, path=str('../graph_data/processed/current_g'), data=False, delimiter='\t')
        WALL_START, CPU_START = time.time(), time.clock()
        p = subprocess.Popen(["/Users/trentonford/Repositories/RWWP/seq/./main"], stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=False, env=my_env)

        (output, err) = p.communicate()

        p_status = p.wait()
        WALL_END, CPU_END = time.time(), time.clock()

        scaling_data_row = pd.DataFrame([["ERDOS_RENYI", G.order(), G.size() ,WALL_END - WALL_START, CPU_END-CPU_START, output]], columns=['TYPE', 'ORDER', 'SIZE', 'WALL', 'CPU', 'RESULT'])
        scaling_data = scaling_data.append(scaling_data_row);

        print ("Command output : ", output)
        print ("Command exit status/return code : ", p_status)
scaling_data.reset_index(inplace=True)
scaling_data.drop(columns='index', inplace=True)
display(scaling_data)
scaling_data.to_pickle('1000_ERDOS_RENYI_Scaling_data')
scaling_data.groupby(['ORDER','SIZE']).agg(['mean', 'max'])
```

```python
import numpy as np
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m-h, m+h
```
